```python
#Check if the matrix is diagonal
import numpy as np
def isDiagonal(mat):
 diag=True
 for i in range(len(mat)):    
    for j in range(len(mat[0])):
     # check for diagonals element
      if i!=j and mat[i][j]!=0:     
            diag=False
            break
 return diag
    
```

```python
img_1d=img[100,:,:]
```

```python
isDiagonal(img_1d)
```

```python
#Check if the matrix is triagonal
def isTriagonal(mat):
    lower=np.allclose(mat, np.tril(mat))
    upper=np.allclose(mat, np.triu(mat))
    return upper,lower
```

```python
isTriagonal(img_1d)
```

```python
#Check if the matrix is identity matrix
def isIdentity(mat):
 iden=True
 for i in range(len(mat)):    
    for j in range(len(mat[0])):
     # check for diagonals element
      if i!=j and mat[i][j]!=0:     
        iden=False
        break
      elif i==j :
        if mat[i][j]!=1 :
         iden=False
         break
 return iden
    
```

```python
isIdentity(img_1d)
```

```python
#Check if the matrix B is an inverse of matrix A
def isInverse(A,B):
    inv=np. linalg. inv(A)
    if inv.all()==B.all() :
        return True
    else :
        return False

```

```python
m= np.array([[6, 1, 1],
              [4, -2, 5],
              [2, 8, 7]])
m_i=np. linalg. inv(m)
isInverse(m,m_i)
```
